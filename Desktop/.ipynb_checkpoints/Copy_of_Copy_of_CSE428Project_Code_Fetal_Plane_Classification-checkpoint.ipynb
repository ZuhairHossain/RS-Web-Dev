```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from tensorflow.python.keras import Sequential
from tensorflow.keras import layers, optimizers, losses, metrics, Model, Input
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import plot_model, to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
folder_link='/content/drive/MyDrive/CSE428_PROJECT_ROOT/'
save_folder='/content/drive/MyDrive/Thesis/Model_Progress/VGG19/'
images_link=folder_link+'DATA/FETAL_PLANES/Images_bmp'
csv_link=folder_link+'DATA/FETAL_PLANES/FETAL_PLANES_DB_data.csv'
model_file = save_folder+'MODEL/vgg19_model_428.h5'
```

```python
data = pd.read_csv(csv_link, sep=';')
data.info()
```

```python
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'Plane' and 'Brain_plane'.
data['Plane']= label_encoder.fit_transform(data['Plane'])

data['Brain_plane']= label_encoder.fit_transform(data['Brain_plane'])
```

```python
limit = 70 #percentage to include

random_index = np.random.randint(0,data.shape[0],limit*data.shape[0]//100)
```

```python
imgdata = []

for img_name in data['Image_name'][random_index]:
  img_path = images_link+'/'+img_name+'.bmp'
 
  if os.path.exists(img_path):
    img_arr=cv2.imread(img_path,0)
    img_arr=np.array(img_arr,dtype='float32')
    img_arr=img_arr/255.0
    imgdata.append(img_arr)
  else:
    print('Did not find image for:',img_name)
X = np.array(imgdata)


labeldata = []

for label in data[['Plane', 'Brain_plane']].values[random_index]:
  labeldata.append(int(str(label[0])+str(label[1])))

Y = np.array(labeldata)

print(type(X),X.shape)
print(type(Y),Y.shape)
```

```python
labelsize = np.unique(Y).shape[0] # 9
X = np.expand_dims(X,axis = 3)  # Expanding the dimension of grayscale images
Y = label_encoder.fit_transform(Y)  # Encoding to 0 to 8
Y = to_categorical(Y, num_classes=labelsize) # One-Hot-Encoding array

print(type(X),X.shape)
print(type(Y),Y.shape)
```

```python
# training on 80% 
train_size=0.8
# split the data in training and remaining dataset 
X_train, X_rem, Y_train, Y_rem = train_test_split(X,Y,train_size=train_size)

# equal valid and test size (10% each of overall data). 
test_size = 0.5
X_valid, X_test, Y_valid, Y_test = train_test_split(X_rem,Y_rem,test_size=test_size)
```

```python
# saving to files, for memory issues
with open(save_folder+'DATA/X_train.npy', 'wb') as f:
    np.save(f, X_train)
with open(save_folder+'DATA/X_test.npy', 'wb') as f:
    np.save(f, X_test)
with open(save_folder+'DATA/X_valid.npy', 'wb') as f:
    np.save(f, X_valid)
with open(save_folder+'DATA/Y_train.npy', 'wb') as f:
    np.save(f, Y_train)
with open(save_folder+'DATA/Y_test.npy', 'wb') as f:
    np.save(f, Y_test)
with open(save_folder+'DATA/Y_valid.npy', 'wb') as f:
    np.save(f, Y_valid)
```

```python
# VGG19 Model

def model_vgg19(inputs, label_count):
  x = Conv2D(64, kernel_size=3, activation="relu", padding="same", name='block1_conv1')(inputs)
  x = Conv2D(64, kernel_size=3, activation="relu", padding="same", name='block1_conv2')(x)
  block_1_output = MaxPooling2D(2, name='block1_pool')(x)

  x = Conv2D(128, kernel_size=3, activation='relu', padding="same", name='block2_conv1')(block_1_output)
  x = Conv2D(128, kernel_size=3, activation='relu', padding="same", name='block2_conv2')(x)
  block_2_output = MaxPooling2D(2, name='block2_pool')(x)

  x = Conv2D(256, kernel_size=3, activation="relu", padding="same", name='block3_conv1')(block_2_output)
  x = Conv2D(256, kernel_size=3, activation="relu", padding="same", name='block3_conv2')(x)
  x = Conv2D(256, kernel_size=3, activation="relu", padding="same", name='block3_conv3')(x)
  x = Conv2D(256, kernel_size=3, activation="relu", padding="same", name='block3_conv4')(x)
  block_3_output = MaxPooling2D(2,name='block3_pool')(x)

  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block4_conv1')(block_3_output)
  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block4_conv2')(x)
  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block4_conv3')(x)
  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block4_conv4')(x)
  block_4_output = MaxPooling2D(2,name='block4_pool')(x)

  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block5_conv1')(block_4_output)
  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block5_conv2')(x)
  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block5_conv3')(x)
  x = Conv2D(512, kernel_size=3, activation="relu", padding="same", name='block5_conv4')(x)
  block_5_output = MaxPooling2D(2,name='block5_pool')(x)

  x = tf.keras.layers.Flatten(name='flatten')(block_5_output)
  x = tf.keras.layers.Dense(4096, activation='relu', name='fc1')(x)
  x = tf.keras.layers.Dense(4096, activation='relu', name='fc2')(x)
  outputs = tf.keras.layers.Dense(label_count, activation="softmax", name='predictions')(x)

  model = tf.keras.Model(inputs=inputs, outputs=outputs, name="vgg19")
  return model

inputs = tf.keras.Input(shape=(224, 224, 1),name='input_1')

model = model_vgg19(inputs, labelsize)
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.RMSprop(1e-4),
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=[
        tf.keras.metrics.CategoricalAccuracy(name='acc'),
    ]
)
model.save(model_file)
```

```python
plot_model(model, show_shapes=True)
```

```python

# model = load_model(model_file)
model_checkpoint = ModelCheckpoint(model_file, monitor="acc",verbose=1, save_best_only=True)
record = model.fit(X_train, Y_train, batch_size=32, epochs=11, validation_data=(X_valid, Y_valid), callbacks=[model_checkpoint])
```

```python
from datetime import datetime,timezone,timedelta
time = datetime.now(tz=timezone(timedelta(hours=6))).strftime("%y_%m_%d_%H_%M_%S")
```

```python
# losses
plt.plot(record.history['loss'], label='train loss')
plt.plot(record.history['val_loss'], label='val loss')
plt.legend()

plt.savefig(folder_link+'MODEL/vgg-loss-428-'+time+'.png')
plt.show()
```

```python
# accuracies
plt.plot(record.history['acc'], label='train acc')
plt.plot(record.history['val_acc'], label='val acc')
plt.legend()

plt.savefig(folder_link+'MODEL/vgg-acc-428-'+time+'.png')
plt.show()
```

```python
model.evaluate(X_test,Y_test,batch_size=32)
```
